### Preguntas y Respuestas


- Reglas del Juego de "Preguntas y Respuestas de Geograf√≠a":

    - El juego consiste en hacer preguntas de geograf√≠a y que el jugador adivine la respuesta correcta.

    - Cada pregunta tiene una respuesta √∫nica y correcta.

    - El juego sigue hasta que el jugador responda incorrectamente 3 de veces o haya respondido 5 preguntas correctamente.


```python
import random

# Diccionario con preguntas y respuestas
preguntas_respuestas = {
    '¬øCu√°l es la capital de Francia?': 'paris',
    '¬øCu√°l es el continente m√°s grande?': 'asia',
    '¬øQu√© pa√≠s tiene la mayor poblaci√≥n del mundo?': 'india',
    '¬øCu√°l es el r√≠o m√°s largo del mundo?': 'amazonas',
    '¬øEn qu√© pa√≠s se encuentra la Torre Eiffel?': 'francia',
    '¬øCu√°l es la capital de Jap√≥n?': 'tokio',
    '¬øCu√°l es el desierto caliente m√°s grande del mundo?': 'sahara',
    '¬øCu√°l es la capital de Argentina?': 'buenosaires',
    '¬øQu√© pa√≠s es conocido como la tierra de los canguros?': 'australia'
}

# Variables para llevar el conteo
conteo_preguntas = 0
respuestas_correctas = 0
respuestas_incorrectas = 0

# Convertimos el diccionario en lista de preguntas
preguntas_respuestas_lista = list(preguntas_respuestas.keys())

# Barajamos las preguntas para que salgan en orden aleatorio sin repetir
random.shuffle(preguntas_respuestas_lista)

print("¬°Bienvenido/a al juego de preguntas! Responde correctamente 5 preguntas para ganar. Puedes fallar hasta 3 veces.")

# Bucle principal
while preguntas_respuestas_lista:
    pregunta_random = preguntas_respuestas_lista.pop(0)  # Tomamos la primera pregunta (ya barajada)
    conteo_preguntas += 1
    print(f"\nPregunta n√∫mero {conteo_preguntas}:")
    
    # Input protegido: evitar respuestas vac√≠as
    while True:
        respuesta_usuario = input(f"{pregunta_random}\n> ").strip().lower().replace(" ", "")
        if respuesta_usuario:
            break
        else:
            print("Por favor, escribe una respuesta v√°lida.")
    
    # Verificamos la respuesta
    if respuesta_usuario == preguntas_respuestas[pregunta_random]:
        respuestas_correctas += 1
        print(f"‚úÖ ¬°Correcto! Total de respuestas correctas: {respuestas_correctas}")
        
        if respuestas_correctas == 5:
            print("\nüéâ ¬°Felicidades, ganaste el juego! üéâ")
            break
        else:
            print("¬°Siguiente pregunta!")
    else:
        respuestas_incorrectas += 1
        print(f"‚ùå Incorrecto. Total de errores: {respuestas_incorrectas}")
        
        if respuestas_incorrectas == 3:
            print("\nüò¢ Has alcanzado el n√∫mero m√°ximo de errores. Fin del juego.")
            break
        else:
            print("¬°No te rindas, sigue intentando!")

# Resumen final
print("\n--- Resumen del juego ---")
print(f"Preguntas respondidas: {conteo_preguntas}")
print(f"Respuestas correctas: {respuestas_correctas}")
print(f"Respuestas incorrectas: {respuestas_incorrectas}")
print("¬°Gracias por jugar! üéÆ")
```

## English

### Questions and Answers

- Rules of the "Geography Quiz Game":

    - The game consists of asking geography-related questions and having the player guess the correct answer.

    - Each question has one unique and correct answer.

    - The game continues until the player either answers incorrectly 3 times or answers 5 questions correctly.

```python
import random

# Dictionary with questions and answers
questions_answers = {
    'What is the capital of France?': 'paris',
    'What is the largest continent?': 'asia',
    'Which country has the largest population?': 'india',
    'What is the longest river in the world?': 'amazonas',
    'In which country is the Eiffel Tower located?': 'france',
    'What is the capital of Japan?': 'tokyo',
    'What is the largest hot desert in the world?': 'sahara',
    'What is the capital of Argentina?': 'buenosaires',
    'Which country is known as the land of kangaroos?': 'australia'
}

# Variables to keep track of progress
question_count = 0
correct_answers = 0
incorrect_answers = 0

# Convert the dictionary into a list of questions
questions_list = list(questions_answers.keys())

# Shuffle the questions to ensure random order without repetition
random.shuffle(questions_list)

print("Welcome to the quiz game! Answer 5 questions correctly to win. You can make up to 3 mistakes.")

# Main game loop
while questions_list:
    random_question = questions_list.pop(0)  # Take the first question (already shuffled)
    question_count += 1
    print(f"\nQuestion number {question_count}:")
    
    # Protected input: avoid empty answers
    while True:
        user_answer = input(f"{random_question}\n> ").strip().lower().replace(" ", "")
        if user_answer:
            break
        else:
            print("Please enter a valid answer.")
    
    # Check the user's answer
    if user_answer == questions_answers[random_question]:
        correct_answers += 1
        print(f"‚úÖ Correct! Total correct answers: {correct_answers}")
        
        if correct_answers == 5:
            print("\nüéâ Congratulations, you won the game! üéâ")
            break
        else:
            print("Let's move on to the next question!")
    else:
        incorrect_answers += 1
        print(f"‚ùå Incorrect. Total mistakes: {incorrect_answers}")
        
        if incorrect_answers == 3:
            print("\nüò¢ You have reached the maximum number of mistakes. Game over.")
            break
        else:
            print("Don't give up, keep trying!")

# Final game summary
print("\n--- Game Summary ---")
print(f"Questions answered: {question_count}")
print(f"Correct answers: {correct_answers}")
print(f"Incorrect answers: {incorrect_answers}")
print("Thanks for playing! üéÆ")
```
