### Reglas del juego de "Piedra, Papel o Tijera"

El objetivo del juego es ser el primero en alcanzar 3 puntos.

Ambas opciones se comparan para determinar qui茅n gana la ronda:

Tijera vence a Papel.
Papel vence a Piedra.
Si ambos jugadores eligen la misma opci贸n, la ronda se considera un empate.

##### Puntuaci贸n:

Cada vez que el jugador gana una ronda, suma 1 punto.

##### Fin del juego:

El juego finaliza cuando uno de los jugadores (el jugador o el ordenador) alcanza los 3 puntos.
Si el jugador gana, se muestra un mensaje de felicitaci贸n. 
Si el ordenador gana, se muestra un mensaje indicando la derrota.

```python
import random  # Importar random para usar random.choice()

# Definir el objetivo de puntos para ganar el juego
obj_puntos = 3
# Inicializar los puntos del usuario y del ordenador
puntos_usuario = 0
puntos_ordenador = 0

# Lista de opciones posibles para jugar
opciones = ['piedra', 'papel', 'tijera']

# Elegir la respuesta del ordenador de forma aleatoria
respuesta_ordenador = random.choice(opciones)

# Bucle principal del juego
while True: 
    # Pedir al jugador que elija una opci贸n
    respuesta_usuario = input('Escoge una opci贸n: piedra, papel o tijera ').lower()
    
    # Validar que la opci贸n ingresada est茅 en las opciones v谩lidas
    if respuesta_usuario not in opciones:
        print('Introduce una de las opciones v谩lidas.')
        respuesta_usuario = input('Escoge una opci贸n v谩lida: piedra, papel o tijera ').lower()

    # Caso de empate
    elif respuesta_usuario == respuesta_ordenador: 
        print(f'Tu respuesta: {respuesta_usuario}')
        print(f'Respuesta ordenador: {respuesta_ordenador}')
        print('Empate')
        print(f'Marcador: usuario: {puntos_usuario} - ordenador: {puntos_ordenador}')
        print('---------------------------------')
        
    # Caso en que gana el ordenador
    elif (
        (respuesta_usuario == 'piedra' and respuesta_ordenador == 'papel') or 
        (respuesta_usuario == 'papel' and respuesta_ordenador == 'tijera') or 
        (respuesta_usuario == 'tijera' and respuesta_ordenador == 'piedra')
    ):
        puntos_ordenador += 1
        print(f'Tu respuesta: {respuesta_usuario}')
        print(f'Respuesta ordenador: {respuesta_ordenador}')
        print('Has perdido esta ronda')
        print(f'Marcador: usuario: {puntos_usuario} - ordenador: {puntos_ordenador}')
        print('---------------------------------')
        
        # Comprobar si el ordenador ha llegado al objetivo de puntos
        if puntos_ordenador == obj_puntos:
            print('Has perdido ')
            break

    # Caso en que gana el usuario
    elif (
        (respuesta_usuario == 'piedra' and respuesta_ordenador == 'tijera') or 
        (respuesta_usuario == 'papel' and respuesta_ordenador == 'piedra') or 
        (respuesta_usuario == 'tijera' and respuesta_ordenador == 'papel')
    ):
        puntos_usuario += 1
        print(f'Tu respuesta: {respuesta_usuario}')
        print(f'Respuesta ordenador: {respuesta_ordenador}')
        print('Has ganado esta ronda')
        print(f'Marcador: usuario: {puntos_usuario} - ordenador: {puntos_ordenador}')
        print('---------------------------------')
        
        # Comprobar si el usuario ha llegado al objetivo de puntos
        if puntos_usuario == obj_puntos: 
            print('隆Felicitaciones! 隆Has ganado! ')
            break
```

### Rules of the "Rock, Paper, Scissors" Game
The player competes against the computer in a series of "rock, paper, scissors" rounds.
The objective of the game is to be the first to reach 3 points.

##### How to Play:
At the start of each round, the computer randomly selects one option among rock, paper, or scissors, and the player chooses one of the three options.
Both choices are then compared to determine the winner of the round:

Rock beats Scissors.

Scissors beats Paper.

Paper beats Rock.

If both the player and the computer choose the same option, the round is considered a tie.

##### Scoring:
Each time the player wins a round, they earn 1 point.

Each time the computer wins a round, it earns 1 point.

If the round is a tie, no points are awarded.

After each round, the current score is displayed, showing the player's and computer's points.
It also indicates whether the player won, lost, or tied the round.

##### End of the Game:
The game ends when either the player or the computer reaches 3 points.
If the player wins, a congratulatory message is displayed.
If the computer wins, a message indicating defeat is shown.

```python
import random  # Import random to use random.choice()

# Define the points objective to win the game
points_goal = 3
# Initialize user and computer points
user_points = 0
computer_points = 0

# List of possible options
options = ['rock', 'paper', 'scissors']

# Randomly choose the computer's response
computer_choice = random.choice(options)

# Main game loop
while True:
    # Ask the player to choose an option
    user_choice = input('Choose an option: rock, paper, or scissors ').lower()
    
    # Validate that the input is among the valid options
    if user_choice not in options:
        print('Please enter a valid option.')
        user_choice = input('Choose a valid option: rock, paper, or scissors ').lower()

    # Case of a tie
    elif user_choice == computer_choice:
        print(f'Your choice: {user_choice}')
        print(f'Computer\'s choice: {computer_choice}')
        print('It\'s a tie!')
        print(f'Score: user: {user_points} - computer: {computer_points}')
        print('---------------------------------')
        
    # Case where the computer wins
    elif (
        (user_choice == 'rock' and computer_choice == 'paper') or 
        (user_choice == 'paper' and computer_choice == 'scissors') or 
        (user_choice == 'scissors' and computer_choice == 'rock')
    ):
        computer_points += 1
        print(f'Your choice: {user_choice}')
        print(f'Computer\'s choice: {computer_choice}')
        print('You lost this round')
        print(f'Score: user: {user_points} - computer: {computer_points}')
        print('---------------------------------')
        
        # Check if the computer reached the points goal
        if computer_points == points_goal:
            print('You lost ')
            break

    # Case where the user wins
    elif (
        (user_choice == 'rock' and computer_choice == 'scissors') or 
        (user_choice == 'paper' and computer_choice == 'rock') or 
        (user_choice == 'scissors' and computer_choice == 'paper')
    ):
        user_points += 1
        print(f'Your choice: {user_choice}')
        print(f'Computer\'s choice: {computer_choice}')
        print('You won this round')
        print(f'Score: user: {user_points} - computer: {computer_points}')
        print('---------------------------------')
        
        # Check if the user reached the points goal
        if user_points == points_goal:
            print('Congratulations! You won! ')
            break
```
