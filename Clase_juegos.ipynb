### Clase incluyendo todos los juegos

```python
import random

class Juegos_clasicos:
    def __init__(self, nombre):
        self.nombre = nombre
    
    def jugadora(self):
        print(f"Hola {self.nombre}, bienvenida a Juegos Cl√°sicos con Python")
    
    def jugar(self):
        print("¬øA qu√© juego quieres jugar?")
        print("Introduce una de las siguientes opciones:")
        print('''        1 --> Si quieres jugar al AHORCADO
        2 --> Si quieres jugar a PIEDRA, PAPEL, TIJERA
        3 --> Si quieres jugar a PREGUNTAS Y RESPUESTAS
        4 --> Si quieres jugar al TRES EN RAYA''')
        
        try:
            opcion = int(input("¬øA qu√© juego quieres jugar?: 1, 2, 3 o 4: "))      
            if opcion == 1:
                self.ahorcado()
            elif opcion == 2:
                self.p_p_t()
            elif opcion == 3:
                self.geografia()
            elif opcion == 4:
                self.tres_en_raya()
            else:
                print("Opci√≥n no v√°lida, introduce un n√∫mero del 1 al 4.")
        except ValueError:
            print("Formato incorrecto, introduce un n√∫mero del 1 al 4.")

    def ahorcado(self):
        print(f"{self.nombre}, ¬øpreparada para jugar al Ahorcado?")
        print("-------------------------------------")
        print("Antes de empezar, te explico las Reglas del Juego")
        print("-------------------------------------")
        print('''Reglas del juego del "Ahorcado":

        - La jugadora intenta adivinar letras para completar la palabra secreta.
        - Si adivina una letra correctamente, se revela en su lugar.
        - Si adivina incorrectamente, se dibuja una parte del cuerpo en la horca.
        - El objetivo es adivinar la palabra antes de completar el dibujo.''')

        dibujos_ahorcado = [
            ''' +---+
                |   |
                    |
                    |
                    |
                    |
                    |
                =========''',
            ''' +---+
                |   |
                O   |
                    |
                    |
                    |
                    |
                =========''',
            ''' +---+
                |   |
                O   |
                |   |
                |   |
                    |
                    |
                =========''',
            ''' +---+
                |   |
                O   |
               /|   |
                |   |
                    |
                    |
                =========''',
            ''' +---+
                |   |
                O   |
               /|\\  |
                |   |
                    |
                    |
                =========''',
            ''' +---+
                |   |
                O   |
               /|\\  |
                |   |
               /    |
                    |
                =========''',
            ''' +---+
                |   |
                O   |
               /|\\  |
                |   |
               / \\  |
                ========='''
        ]

        palabras_secretas = ["oro", "jupiter", "coche"]
        palabra_secreta = random.choice(palabras_secretas)

        vidas_usuario = 6
        palabra_oculta = ["_"] * len(palabra_secreta)
        letras_usadas = []

        print("\nüéâ ¬°Bienvenida al Ahorcado!")
        print(dibujos_ahorcado[0])
        print("Palabra oculta: " + " ".join(palabra_oculta))
        input("\nPresiona ENTER para comenzar...")

        while True:
            letra_usuario = input("\nIntroduce una letra: ").lower().strip()
            print('********************************************')

            if not letra_usuario.isalpha() or len(letra_usuario) != 1:
                print("‚ùó Introduce solo una letra v√°lida.")
                continue

            if letra_usuario in letras_usadas:
                print(f"‚ö†Ô∏è Ya has usado '{letra_usuario}'.")
                continue

            letras_usadas.append(letra_usuario)

            if letra_usuario in palabra_secreta:
                print(f"‚úÖ ¬°Correcto! '{letra_usuario}' est√° en la palabra.")
                for i, letra_correcta in enumerate(palabra_secreta):
                    if letra_correcta == letra_usuario:
                        palabra_oculta[i] = letra_usuario

                if "_" not in palabra_oculta:
                    print("\nüéâ ¬°Ganaste! La palabra era:", palabra_secreta.upper())
                    break
            else:
                vidas_usuario -= 1
                print(f"‚ùå '{letra_usuario}' no est√° en la palabra.")
                print(f"Vidas restantes: {'‚ù§Ô∏è' * vidas_usuario}")

            vidas_perdidas = 6 - vidas_usuario
            print(dibujos_ahorcado[vidas_perdidas])
            print("Palabra actual: " + " ".join(palabra_oculta))

            if vidas_usuario == 0:
                print("\nüíÄ ¬°Perdiste! La palabra era:", palabra_secreta.upper())
                break

    def p_p_t(self):
        print(f"Hola {self.nombre}, ¬øpreparada para jugar a Piedra, Papel o Tijera?")
        print("-------------------------------------")
        print('''Reglas del juego:

        - Juegas contra el ordenador.
        - Piedra vence a Tijera, Tijera vence a Papel, Papel vence a Piedra.
        - Ganas si llegas a 3 puntos primero.''')

        obj_puntos = 3
        puntos_usuario = 0
        puntos_ordenador = 0
        opciones = ['piedra', 'papel', 'tijera']

        while True:
            respuesta_usuario = input('Escoge piedra, papel o tijera: ').lower()
            respuesta_ordenador = random.choice(opciones)

            if respuesta_usuario not in opciones:
                print("Opci√≥n inv√°lida. Intenta de nuevo.")
                continue

            print(f'Tu respuesta: {respuesta_usuario}')
            print(f'Respuesta ordenador: {respuesta_ordenador}')

            if respuesta_usuario == respuesta_ordenador:
                print('Empate')
            elif (respuesta_usuario == 'piedra' and respuesta_ordenador == 'tijera') or \
                 (respuesta_usuario == 'papel' and respuesta_ordenador == 'piedra') or \
                 (respuesta_usuario == 'tijera' and respuesta_ordenador == 'papel'):
                puntos_usuario += 1
                print('¬°Ganaste esta ronda!')
            else:
                puntos_ordenador += 1
                print('Perdiste esta ronda.')

            print(f'Marcador: usuario {puntos_usuario} - ordenador {puntos_ordenador}')
            print('---------------------------------')

            if puntos_usuario == obj_puntos:
                print('¬°Felicidades, ganaste el juego! üéâ')
                break
            elif puntos_ordenador == obj_puntos:
                print('Has perdido el juego. üòû')
                break

    def geografia(self):
        print(f"Hola {self.nombre}, ¬°a poner a prueba tus conocimientos de geograf√≠a!")
        print("-------------------------------------")
        print('''Reglas:

        - Adivina las respuestas correctas de preguntas de geograf√≠a.
        - Tienes 2 intentos por pregunta.
        - El juego termina si fallas 3 veces o aciertas 5 veces.''')

        preguntas_respuestas = {
            '¬øCu√°l es la capital de Francia?': 'paris',
            '¬øCu√°l es el continente m√°s grande?': 'asia',
            '¬øQu√© pa√≠s tiene la mayor poblaci√≥n del mundo?': 'india',
            '¬øCu√°l es el r√≠o m√°s largo del mundo?': 'amazonas',
            '¬øEn qu√© pa√≠s se encuentra la Torre Eiffel?': 'francia',
            '¬øCu√°l es la capital de Jap√≥n?': 'tokio',
            '¬øCu√°l es el desierto caliente m√°s grande del mundo?': 'sahara',
            '¬øCu√°l es la capital de Argentina?': 'buenosaires',
            '¬øQu√© pa√≠s es conocido como la tierra de los canguros?': 'australia'
        }

        preguntas = list(preguntas_respuestas.keys())
        random.shuffle(preguntas)

        correctas = 0
        incorrectas = 0

        print("\n¬°Empieza el juego!")

        while preguntas:
            pregunta = preguntas.pop(0)
            print(f"\nPregunta: {pregunta}")

            intento = input("> ").strip().lower().replace(" ", "")

            if intento == preguntas_respuestas[pregunta]:
                correctas += 1
                print("‚úÖ Correcto.")
            else:
                incorrectas += 1
                print("‚ùå Incorrecto.")

            print(f"Correctas: {correctas} | Incorrectas: {incorrectas}")

            if correctas == 5:
                print("\nüéâ ¬°Ganaste el juego de geograf√≠a!")
                break
            if incorrectas == 3:
                print("\nüò¢ Has perdido el juego de geograf√≠a.")
                break

    def tres_en_raya(self):
        print(f"Hola {self.nombre}, ¬ølista para el Tres en Raya?")
        print("-------------------------------------")
        print('''Reglas:

        - Dos jugadoras (X y O) alternan turnos en un tablero 3x3.
        - El objetivo es formar una l√≠nea horizontal, vertical o diagonal con el mismo s√≠mbolo.
        - El juego termina con victoria o empate.''')

        huecos_tablero = {
            'a1': '-', 'b1': '-', 'c1': '-',
            'a2': '-', 'b2': '-', 'c2': '-',
            'a3': '-', 'b3': '-', 'c3': '-'
        }

        def imprimir_tablero():
            tablero = f'''
            +-----+-----+-----+
            |  a  |  b  |  c  |
    _______________________
    | 1 |  {huecos_tablero['a1']}  |  {huecos_tablero['b1']}  |  {huecos_tablero['c1']}  |
    _______________________
    | 2 |  {huecos_tablero['a2']}  |  {huecos_tablero['b2']}  |  {huecos_tablero['c2']}  |
    _______________________
    | 3 |  {huecos_tablero['a3']}  |  {huecos_tablero['b3']}  |  {huecos_tablero['c3']}  |
            +-----+-----+-----+
            '''
            print(tablero)

        def comprobar_ganador():
            combinaciones = [
                ['a1', 'b1', 'c1'], ['a2', 'b2', 'c2'], ['a3', 'b3', 'c3'],
                ['a1', 'a2', 'a3'], ['b1', 'b2', 'b3'], ['c1', 'c2', 'c3'],
                ['a1', 'b2', 'c3'], ['a3', 'b2', 'c1']
            ]
            for combo in combinaciones:
                if huecos_tablero[combo[0]] == huecos_tablero[combo[1]] == huecos_tablero[combo[2]] and huecos_tablero[combo[0]] != '-':
                    return huecos_tablero[combo[0]]
            if '-' not in huecos_tablero.values():
                return 'E'
            return None

        turno = 'X'
        while True:
            imprimir_tablero()
            coordenada = input(f"Turno {turno}. Introduce coordenada (ej: a1): ").lower()

            if coordenada in huecos_tablero and huecos_tablero[coordenada] == '-':
                huecos_tablero[coordenada] = turno
                resultado = comprobar_ganador()

                if resultado == 'X':
                    imprimir_tablero()
                    print("¬°Gan√≥ el jugador X! üéâ")
                    break
                elif resultado == 'O':
                    imprimir_tablero()
                    print("¬°Gan√≥ el jugador O! üéâ")
                    break
                elif resultado == 'E':
                    imprimir_tablero()
                    print("¬°Empate! üòÆ")
                    break

                turno = 'O' if turno == 'X' else 'X'
            else:
                print("Coordenada no v√°lida o ocupada. Intenta de nuevo.")
```
